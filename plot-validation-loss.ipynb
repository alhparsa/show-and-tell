```python
from __future__ import division
import matplotlib.pyplot as plt
import utils
import numpy as np
%matplotlib inline
```

```python
def plotter(losses_lstm,losses_gru,losses_elman,title,output_filename,xlabel='',ylabel='loss'):
    #losses_val_1 = loss_dict_1['losses_val']
    #losses_val_2 = loss_dict_2['losses_val']

    fig = plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
    
    # plotting training and val together requires shortening losses_train to 76 * 124
    y1 = np.array(losses_elman)
    y2 = np.array(losses_lstm)
    y3 = np.array(losses_gru)
    plt.plot(y1,'r')
    plt.plot(y2,'g')
    plt.plot(y3,'b')
    
    plt.gca().set_ylim([1.8,3])

    # LSTM memory size
    plt.title(title)
    plt.legend(('basic_rnn', 'lstm', 'gru'),loc='upper right')
    plt.xlabel(xlabel, fontsize=18)
    plt.ylabel(ylabel, fontsize=16)
    plt.tight_layout()
    
    fig.savefig(output_filename)

    
    
files = {"elman512" : "losses/losses_val2014_elman_e512_h512.pkl",
         "elman1024" : "losses/losses_val2014_elman_e512_h1024.pkl",
         "gru512" : "losses/losses_val2014_gru_e512_h512.pkl",
         "gru1024" : "losses/losses_val2014_gru_e512_h1024.pkl",
         "lstm512" : "losses/losses_val2014_lstm_e512_h512.pkl",
         "lstm1024" : "losses/losses_val2014_lstm_e512_h1024.pkl"}
    

elman512_trainval = utils.unpickle(files["elman512"])
elman512_train = elman512_trainval["losses_train"]
elman512_val = elman512_trainval["losses_val"]

elman1024_trainval = utils.unpickle(files["elman1024"])
elman1024_train = elman1024_trainval["losses_train"]
elman1024_val = elman1024_trainval["losses_val"]

gru512_trainval = utils.unpickle(files["gru512"])
gru512_train = gru512_trainval["losses_train"]
gru512_val = gru512_trainval["losses_val"]

gru1024_trainval = utils.unpickle(files["gru1024"])
gru1024_train = gru1024_trainval["losses_train"]
gru1024_val = gru1024_trainval["losses_val"]

lstm512_trainval = utils.unpickle(files["lstm512"])
lstm512_train = lstm512_trainval["losses_train"]
lstm512_val = lstm512_trainval["losses_val"]

lstm1024_trainval = utils.unpickle(files["lstm1024"])
lstm1024_train = lstm1024_trainval["losses_train"]
lstm1024_val = lstm1024_trainval["losses_val"]


```

```python
plotter(lstm512_train,
        gru512_train,
        elman512_train,
        'Loss, train set, rnn_state_size=512',
        'plot_train_512.png',
        xlabel='steps (batches)')

plotter(lstm512_val,
        gru512_val,
        np.array(elman512_val),
        'Loss, validation set, rnn_state_size=512',
        'plot_val_512.png',
        xlabel='complete iterations over validation set',
        ylabel='average loss')

plotter(lstm1024_train,
        gru1024_train,
        elman1024_train,
        'Loss, train set, rnn_state_size=1024',
        'plot_train_1024.png',
        xlabel='steps (batches)')

plotter(lstm1024_val,
        gru1024_val,
        np.array(elman1024_val),
        'Loss, validation set, rnn_state_size=1024',
        'plot_val_1024.png',
        xlabel='complete iterations over validation set',
        ylabel='average loss')
```
